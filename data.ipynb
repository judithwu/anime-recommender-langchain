```python
import pandas as pd
```

```python
anime = pd.read_csv("anime.csv")
anime.head()
```

```python
#create new column with links to the anime 
anime["link"] = "https://myanimelist.net/anime/" + anime["MAL_ID"].astype(str)
anime.head()
```

```python
#only keep name, genres, synopsis, and link columns
anime = anime.drop(["MAL_ID", "Score"], axis=1)
anime.head()
```

```python
#create a page content column
anime["page_content"] = "Title: " + anime["Name"] + "\n" + \
                        "Genre: " + anime["Genres"] + "\n" + \
                        "Synopsis: " + anime["sypnopsis"]
#only keep page content and link columns
anime = anime.drop(["Name", "Genres", "sypnopsis"], axis=1)
anime.head()
```

```python
import pickle

#pickle dataframe to file
with open('dataframe.pkl', 'wb') as file:
    pickle.dump(anime, file)
```
